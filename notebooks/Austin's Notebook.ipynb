```python
def mean_spacings(series_list):
    x_dim = []
    y_dim = [] 
    z_dim = []
    # sitk read all flair images and rotate
    for i in tqdm(series_list):
        # read sitk image 
        image = read_dicom_series(i)
        # rotate image
        image = rotate_image(image)
        # get spacings 
        sx, sy, sz = image.GetSpacing()

        x_dim.append(sx)
        y_dim.append(sy)
        z_dim.append(sz)
    return np.mean(x_dim), np.mean(y_dim), np.mean(z_dim)
    
```

```python
def get_images(img_dir):
    patients = [img_dir + p for p in next(os.walk(img_dir))[1]]
    images = images = {
        "flair" : [p + "/FLAIR.nii" for p in patients],
        "t1w": [p + "/T1w.nii" for p in patients],
        "t1wce": [p + "/T1wCE.nii" for p in patients],
        "t2w": [p + "/T2w.nii" for p in patients]
    }
    return images
```

```python
def itk_rescale(images, post_is=np.asarray([100,100,100])):
    for i in tqdm(images):
        pre_img = sitk.ReadImage(i, imageIO="NiftiImageIO")
        # get constants 
        pre_vs = np.asarray(pre_img.GetSpacing())
        pre_is = np.asarray(pre_img.GetSize())
        
        # rescale post voxel spacing 
        post_vs = pre_is/post_is * pre_vs
        
        post_img = sitk.Resample(
            image1 = pre_img,
            size = post_is.tolist(),
            transform = sitk.Transform(),
            interpolator = sitk.sitkBSpline,
            outputOrigin = pre_img.GetOrigin(),
            outputSpacing = post_vs.tolist(),
            outputDirection = pre_img.GetDirection(),
            defaultPixelValue = 0.0,
            outputPixelType = pre_img.GetPixelID()
        )
        
        # get patient_id and image type 
        pid = os.path.basename(os.path.dirname(i))
        im_type = os.path.basename(i)
        
        # construct output path
        out_path = "./processed/train/" + pid + "/"
        safe_make_dir("./processed/")
        safe_make_dir("./train/")
        safe_make_dir(out_path)
        out_file = out_path + im_type + ".nii"
        
        
        sitk.WriteImage(post_img, out_file, imageIO="NiftiImageIO")
        
```
